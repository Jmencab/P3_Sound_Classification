# Average Correlation Finder

This file goes through each of the examples in the training set and correlaties them to test data to find the their average correaltion. The classification is then asssigned based on highest average correlation.

```python
import pandas as pd
import scipy.signal as sig
import numpy as np
```

```python
# Load in train data
whole_df = pd.read_csv("train.csv", nrows=500)
```

```python
whole_df.head
```

```python
classes = whole_df.iloc[:,-1]
classes
```

```python
# Load in test data
test_df = pd.read_csv("test.csv", nrows=500)
```

```python
test_df.shape
```

```python
# Group train data by class
df_groups = whole_df.groupby(whole_df.iloc[:,-1])
predictions = np.zeros(len(test_df))
```

```python
for name,group in df_groups:
    print name
    print group.shape
```

```python

group_num = len(df_groups)
# For each row i, set predictions[i] to greatest average correlation
for index, row in test_df.iterrows():
    corr_vector = np.zeros(group_num)
    group_ind = 0
    for _, group in df_groups:
        avg_corr = 0
        for _,signal in group.iterrows():
            avg_corr += sig.correlate(row, signal)
        avg_corr.shape
        corr_vector[group_ind] = np.sum(avg_corr /len(group))
        group_ind += 1
    predictions[index] = np.argmax(corr_vector)
```
